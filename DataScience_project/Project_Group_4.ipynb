```python
import pandas as pd
sExcelFile="data1-5.xlsx"
df1 = pd.read_excel(sExcelFile,sheet_name='Sheet1')
df1
```

Now we can see the details of given data1-5, Now we should use data 10 and then 11 as a complemntary.[df2,df3]

```python
sExcelFile="data10.xlsx"
df2 = pd.read_excel(sExcelFile,sheet_name='Sheet1')
df2
```

```python
sExcelFile="data11.xlsx"
df3 = pd.read_excel(sExcelFile,sheet_name='Sheet1')
df3
```

## Data Gathering

Here we need to merge tables from df1, df2 and df3. Because df1 has dataset of class1-5 and df2 has dataset for class6, and df3 has dataset for class7.
We can also use contatenation or append ; merge or join but we are using merge here. [df=df1+df2; dff=df1+df2+df3]

We used Student_ID to merge

```python
df=df1.merge(df2,on='Student_ID',how='right')
dff=df.merge(df3,on='Student_ID',how='right')
dff
```

## Data pre-processing 


We need to extract the numbers as the features we used in  from class1-class 5, class6 and class7

```python
a1=dff.class1.str.extract('(\d+)')
a2=dff.class2.str.extract('(\d+)')
a3=dff.class3.str.extract('(\d+)')
a4=dff.class4.str.extract('(\d+)')
a5=dff.class5.str.extract('(\d+)')
a6=dff.class6.str.extract('(\d+)')
a7=dff.class7.str.extract('(\d+)')
b = pd.concat([a1,a2,a3,a4,a5,a6,a7],axis=1)
b
```


Here we can see some missing values as absent or NAN, so we can fill them by putting 10 as a value in all.

```python
b=b.fillna(10);b
```

##  Clustering
Here we can use the package sklearn, where the function kmeans is a clustering method, and we will be clustering we attendance as 4 grades A B C D

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans =kmeans.fit(b)
y_kmeans = kmeans.predict(b)
y_kmeans
```

##  Visulazation
Visulaze with hist graph and box graph

```python
import matplotlib.pyplot as plt 
plt.boxplot(y_kmeans)
plt.show()
```

```python
from scipy.stats import norm
fig,ax = plt.subplots()
plt.rcParams['font.family'] = ['SimHei'] 
n,bins,patches = ax.hist(y_kmeans,density = 1)
mu = 1
sigma = 0.4
y=norm.pdf(bins,mu,sigma)
ax.plot(bins,y,'--')
fig.tight_layout()
plt.show()
```
