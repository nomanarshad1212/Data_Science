## 4.  Clustering
You can use the package sklearn, the function kmeans is a clustering method, and you will clusting your attendance as 4 grades A B C D

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

## 5. Visulazation

```python
import matplotlib.pyplot as plt 
```

```python
from scipy.stats import norm
fig,ax = plt.subplots()
plt.rcParams['font.family'] = ['SimHei'] 
n,bins,patches = ax.hist(y_kmeans,density = 1)
mu = 1
sigma = 0.4
y=norm.pdf(bins,mu,sigma)
ax.plot(bins,y,'--')
fig.tight_layout()
plt.show()
```
